```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time

# Указываем путь к драйверу
driver_path = r'c:\Users\Dmitry Kh\Documents\webdriver\chromedriver.exe'

# Инициализируем опции
options = webdriver.ChromeOptions()
options.add_argument(f'--webdriver={driver_path}')

# Инициализируем веб-драйвер с опциями
driver = webdriver.Chrome(options=options)

# Открываем страницу
url = 'https://www.bukalapak.com/c/sepeda/fullbike/sepeda-listrik'
driver.get(url)

# Функция для получения информации о карточках товаров
def process_product_card(card):
    name = card.find_element(By.CLASS_NAME, 'bl-text--ellipsis__2').get_attribute('title')
    price = card.find_element(By.CLASS_NAME, 'bl-product-card-new__price').get_attribute('title')
    
    try:
        original_price = card.find_element(By.CLASS_NAME, 'bl-product-card-new__original-price').text
    except NoSuchElementException:
        original_price = None
    
    sold_count = card.find_element(By.CLASS_NAME, 'bl-product-card-new__sold-count').text
    store_location = card.find_element(By.CLASS_NAME, 'bl-product-card-new__store-location').get_attribute('title')
    
    return {'name': name, 'price': price, 'original_price': original_price, 'sold_count': sold_count, 'store_location': store_location}

# Функция для прокрутки страницы до конца
def scroll_to_bottom():
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Подождем, чтобы данные успели подгрузиться

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Получаем информацию со всех страниц
all_results = []

# Прокручиваем страницу до конца
scroll_to_bottom()

# Обрабатываем карточки товаров
cards = driver.find_elements(By.XPATH, '//div[@class="bl-flex-item mb-8" and contains(@style, "width: 20%")]')

for card in cards:
    product_info = process_product_card(card)
    all_results.append(product_info)

# Закрываем веб-драйвер
driver.quit()

# Выводим результаты
for result in all_results:
    print(result)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, StaleElementReferenceException
import time

# Указываем путь к драйверу
driver_path = r'c:\Users\Dmitry Kh\Documents\webdriver\chromedriver.exe'

# Инициализируем опции
options = webdriver.ChromeOptions()
options.add_argument(f'--webdriver={driver_path}')

# Инициализируем веб-драйвер с опциями
driver = webdriver.Chrome(options=options)

# Функция для получения информации о карточках товаров
def process_product_card(card):
    name = card.find_element(By.CLASS_NAME, 'bl-text--ellipsis__2').get_attribute('title')
    price = card.find_element(By.CLASS_NAME, 'bl-product-card-new__price').get_attribute('title')
    
    try:
        original_price = card.find_element(By.CLASS_NAME, 'bl-product-card-new__original-price').text
    except NoSuchElementException:
        original_price = None
    
    sold_count = card.find_element(By.CLASS_NAME, 'bl-product-card-new__sold-count').text
    
    # Добавляем проверку наличия элемента 'bl-product-card-new__store-location'
    try:
        store_location_element = card.find_element(By.CLASS_NAME, 'bl-product-card-new__store-location')
        store_location = store_location_element.get_attribute('title')
    except NoSuchElementException:
        store_location = None
    
    return {'name': name, 'price': price, 'original_price': original_price, 'sold_count': sold_count, 'store_location': store_location}

# Функция для клика по кнопке "Halaman selanjutnya"
def click_next_page():
    try:
        next_button = driver.find_element(By.XPATH, '//span[@class="bl-ghostblock-pagination__next"]')
        next_button.click()
    except NoSuchElementException:
        print("Кнопка 'Halaman selanjutnya' не найдена")
    except StaleElementReferenceException:
        print("Элемент 'Halaman selanjutnya' устарел и не может быть использован")


# Функция для прокрутки страницы вниз с использованием ActionChains
def scroll_down():
    actions = ActionChains(driver)
    actions.send_keys(Keys.PAGE_DOWN).perform()
    time.sleep(2)  # Подождем, чтобы данные успели подгрузиться

# Получаем информацию со всех страниц
all_results = []

# Прокручиваем страницы
while True:
    cards = driver.find_elements(By.XPATH, '//div[@class="bl-flex-item mb-8" and contains(@style, "width: 20%")]')
    

    for card in cards:
        product_info = process_product_card(card)
        all_results.append(product_info)

    # Прокручиваем страницу вниз
    scroll_down()

    # Переходим на следующую страницу
    click_next_page()

    # Ждем, чтобы данные успели подгрузиться
    time.sleep(2)

    # Проверяем, достигли ли конца страниц
    try:
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[@class="bl-flex-item mb-8" and contains(@style, "width: 20%")]')))
    except TimeoutException:
        break

# Закрываем веб-драйвер
driver.quit()

# Выводим результаты
for result in all_results:
    print(result)
```
