```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.common.exceptions import NoSuchElementException

# Указываем путь к chromedriver
driver_path = r'C:\Users\Dmitry Kh\Мой диск\Работа\Эко байки Индонезия\Аналитика\Запросы\chromedriver-win64\chromedriver.exe'

# Указываем URL страницы с товарами
url = 'https://www.bukalapak.com/c/sepeda/fullbike/sepeda-lipat'

# Создаем экземпляр сервиса Chrome
service = ChromeService(executable_path=driver_path)

# Создаем экземпляр веб-драйвера
driver = webdriver.Chrome(service=service)

# Загружаем страницу
driver.get(url)

# Получаем список товаров
products = driver.find_elements(By.CLASS_NAME, 'product-card')

# Открываем файл для записи
with open('output.txt', 'w', encoding='utf-8') as file:
    # Проходим по каждому товару и извлекаем информацию
    for product in products:
        try:
            # Наименование товара
            nama_element = product.find_element(By.CLASS_NAME, 'product-title')
            nama = nama_element.text if nama_element else "Наименование не найдено"

            # Цена
            harga_element = product.find_element(By.CLASS_NAME, 'product-price')
            harga = harga_element.text if harga_element else "Цена не найдена"

            # Остаток на складе
            stok_element = product.find_element(By.CLASS_NAME, 'u-txt--base.u-fg--red-brand')
            stok = stok_element.text if stok_element else "Остаток не найден"

            # Местонахождение продавца
            lokasi_element = product.find_element(By.CLASS_NAME, 'c-seller__city.u-mrgn-bottom--2')
            lokasi = lokasi_element.text if lokasi_element else "Местонахождение не найдено"

            # Состояние товара
            kondisi_element = product.find_element(By.CLASS_NAME, 'c-information__subtitle')
            kondisi = kondisi_element.text if kondisi_element else "Состояние не найдено"

            # Категория
            kategori_element = product.find_element(By.CLASS_NAME, 'u-align-left.u-txt--normal')
            kategori = kategori_element.text if kategori_element else "Категория не найдена"

            # Масса товара
            berat_element = product.find_element(By.XPATH, '//th[contains(text(), "Berat")]/following-sibling::td')
            berat_text = berat_element.text if berat_element else None

            # Если текст содержит "kilogram", извлекаем числовое значение
            if berat_text and 'kilogram' in berat_text:
                berat = berat_text.split(' ')[0]
            else:
                berat = "Масса товара не указана"

            # Местный или импорт
            asal_element = product.find_element(By.XPATH, '//th[contains(text(), "Asal Barang")]/following-sibling::td')
            asal = asal_element.text if asal_element else "Местный или импорт не указан"

            # Записываем информацию в файл
            file.write("Наименование: {}\n".format(nama))
            file.write("Цена: {}\n".format(harga))
            file.write("Остаток на складе: {}\n".format(stok))
            file.write("Местонахождение продавца: {}\n".format(lokasi))
            file.write("Состояние: {}\n".format(kondisi))
            file.write("Категория: {}\n".format(kategori))
            file.write("Масса: {}\n".format(berat))
            file.write("Местный или импорт: {}\n".format(asal))
            file.write("\n---\n")

        except NoSuchElementException as e:
            print("Ошибка при парсинге товара:", e)

# Закрываем браузер
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.bukalapak.com/c/sepeda/fullbike/sepeda-lipat'
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Предположим, что каждая карточка товара находится в элементе с классом "bl-product-card-new__wrapper"
    product_cards = soup.find_all(class_='bl-product-card-new__wrapper')

    for card in product_cards:
        # Получаем наименование товара из тега <p> с классом "bl-text bl-text--body-14"
        product_name_tag = card.find('p', class_='bl-text bl-text--body-14')
        product_name = product_name_tag.find('a').text.strip() if product_name_tag else 'N/A'

        # Получаем цену товара из соответствующего элемента
        product_price = card.find(class_='bl-text-red').text.strip()

        print(f'Product: {product_name}, Price: {product_price}')
else:
    print(f'Error: {response.status_code}')
```
